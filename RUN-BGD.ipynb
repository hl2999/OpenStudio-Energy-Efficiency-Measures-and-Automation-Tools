```python
def reorder_ecms(osw_paths):
    osw_paths = natsort.natsorted(osw_paths)
    for n, osw_path in enumerate(osw_paths):

        if 'proposed' in osw_path.lower():
            osw_paths.pop(n)
            osw_paths.append(osw_path)
    
    return osw_paths
            
```

```python
def run_all(osw_paths):    
    for osw_file in osw_paths:
        start_time = time.time()
        print(osw_file)
        process = subprocess.Popen(["openstudio.exe", "run", "-w", osw_file], stdout=subprocess.PIPE, stderr=subprocess.PIPE, universal_newlines=True)
        output, errors = process.communicate()
        elapsed_time = time.time() - start_time
        print(elapsed_time)
```

```python
import os

directory = os.getcwd()
output = []
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith("eplusout.sql"):
            output.append(os.path.join(root, file))
            print(os.path.join(root, file))
```

```python
import os
import sqlite3
import pathlib
import pandas as pd

def get_sql_uri(sql_path):
    """
    Set the URI (read-only) to the EnergyPlus SQL from the SQL file path

    Parameters
    ----------
    sql_path : str
        Path to the sql file.

    """
    if not os.path.isfile(sql_path):
        raise IOError("No such file or directory: '%s'" % sql_path)

    abs_sql_path = os.path.abspath(sql_path)
    sql_uri = '{}?mode=ro'.format(pathlib.Path(abs_sql_path).as_uri())
    print('SQL URI set to: {}'.format(sql_uri))

    return sql_uri


def get_annual_energy_by_fuel_and_enduse(sql_path):
    """
    Queries SQL file and returns the ABUPS' End Uses table

    Parameters
    ----------
    * sql_path (str): path to the SQL file on disk.

    Returns
    -------
    df_annual: pd.DataFrame
        Annual End Use table
        index = 'EndUse'
        columns = ['FuelType','Units']
    """

    # RowName = '#{end_use}'
    # ColumnName='#{fuel_type}'
    annual_end_use_query = """SELECT RowName, ColumnName, Units, Value
        FROM TabularDataWithStrings
        WHERE ReportName='AnnualBuildingUtilityPerformanceSummary'
        AND ReportForString='Entire Facility'
        AND TableName='End Uses'
    """

    sql_uri = get_sql_uri(sql_path)

    with sqlite3.connect(sql_uri, uri=True) as con:
        df_annual = pd.read_sql(annual_end_use_query, con=con)

    # Convert Value to Float
    df_annual['Value'] = pd.to_numeric(df_annual['Value'])

    df_annual = df_annual.set_index(['RowName',
                                     'ColumnName',
                                     'Units'])['Value'].unstack([1, 2])
    df_annual.index.name = 'EndUse'
    df_annual.columns.names = ['FuelType', 'Units']

    end_use_order = ['Heating', 'Cooling',
                     'Interior Lighting', 'Exterior Lighting',
                     'Interior Equipment', 'Exterior Equipment',
                     'Fans', 'Pumps', 'Heat Rejection', 'Humidification',
                     'Heat Recovery', 'Water Systems',
                     'Refrigeration', 'Generators', 'Total End Uses']

    col_order = ['Electricity', 'Natural Gas', 'Additional Fuel',
                 'District Cooling', 'District Heating', 'Water']

    df_annual = df_annual[col_order].loc[end_use_order]

    return df_annual
```

```python
def get_ecm_num(sql_file):
    return sql_file[114:120]
```

```python
sql_file[114:120]
```

```python
dfs = []
for sql_file in output[:-1]:
    df = get_annual_energy_by_fuel_and_enduse(sql_path=sql_file)
#     df.loc['Total End Uses']= df.sum()
    
    df = df.T
    df['ECM'] = get_ecm_num(sql_file)
    df = df.set_index(['ECM'], append=True)
    dfs.append(df)

df_annual = pd.concat(dfs)
```

```python
df_annual = df_annual[(df_annual.T != 0).any()] # remove row with all 0
df_annual = df_annual.loc[:, (df != 0).any(axis=0)] # remove column with all 0
```

```python
df_annual
```

```python
# define cost per unit
elec_per_kwh = 0.19
gas_per_therm = 1.32
water_per_kgal = 7

price_dic = {'$/kWh Electricity':elec_per_kwh, '$/Therm Natural Gas':gas_per_therm, '$/ 1000 Gallons Water':water_per_kgal}
df_price_info = pd.Series(price_dic)

costs = []

for index, row in df_annual.iterrows():
    
    if index[0] == 'Electricity':
        cost = 277.778 * elec_per_kwh * row['Total End Uses']
    if index[0] == 'Natural Gas':
        cost = 9.48 * gas_per_therm * row['Total End Uses']
    if index[0] == 'Water':
        cost = 264.172 * water_per_kgal * row['Total End Uses'] / 1000
#         cost = 0
    
    costs.append(cost)
    
df_annual['Cost'] = costs
```

```python
df_elec = df_annual.loc['Electricity']
df_gas = df_annual.loc['Natural Gas']
df_water = df_annual.loc['Water']
```

```python
df_cost = df_annual.groupby(by='ECM').sum()[['Cost']]
```

```python
df_cost
```

```python
savings = []
for n in range(len(df_cost)):
    
    if n == 0:
        saving = 0 # baseline
    else:
        saving = df_cost.iloc[n-1]['Cost'] - df_cost.iloc[n]['Cost']
     
    savings.append(saving)
    
df_cost['cost savings ($)'] = savings
```

```python
df_cost['% savings'] = df_cost['cost savings ($)']/df_cost.iloc[0]['Cost']
```

```python

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('ecm.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df_annual.to_excel(writer, sheet_name='annual summary')
df_cost.to_excel(writer, sheet_name='cost summary')
df_elec.to_excel(writer, sheet_name='elec')
df_gas.to_excel(writer, sheet_name='gas')
df_water.to_excel(writer, sheet_name='water')
df_price_info.to_excel(writer, sheet_name='price info', header=False)
# Close the Pandas Excel writer and output the Excel file.


writer.save()
```
